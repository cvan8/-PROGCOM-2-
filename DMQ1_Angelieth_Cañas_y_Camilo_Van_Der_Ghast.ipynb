###Resumen

Mediante el desarrollo de esta práctica experimental se busca comprender
analíticamente la variación que sufre el pH de una solución debido al aumento de la concentración de equivalentes de base añadidos al realizar un proceso de titulación entre un ácido y una base; que dependiendo de su naturaleza,se pueden contemplar dos situaciones como la titulación de un ácido débil con una base fuerte, como es el caso del ácido acético junto con el Hidróxido de Sodio y
finalmente, el caso de la titulación del aminoácido de la glicina con la misma base. De esta manera, se determinó la curva de titulación donde se representan dichos valores de pH obtenidos a través del uso de un pHmetro y analizando visualmente dichas variaciones mediante el método colorimétrico; lo cual permitió analizar cuando una solución se puede considerar de tipo amortiguadora o tampón al no presentarse cambios bruscos en el pH y así, establecer la importancia biológica de este tipo de soluciones. 


###Introducción

Una solución se define como una mezcla homogénea entre dos o más sustancias, donde de acuerdo a sus propiedades se establece un soluto y un solvente. Mediante el proceso de titulación que es un método de análisis químico cuantitativo se puede determinar en el laboratorio la concentración de una sustancia desconocida, ya que consiste en ir añadiendo un volumen medido de una disolución de concentración, mientras se mide de forma paralela a la adición del
volumen el pH. En esta práctica de laboratorio se observa el comportamiento de titulaciones de tipo acido-base, las cuales consisten de una disolución ácida o básica desconocida, mediante una neutralización controlada, al añadir más solución volumétrica, ya que la solución de prueba se volverá más ácido, encontrará el punto de equilibrio o será más básico. Posteriormente, se realiza la curva de titulación de la solución, considerando los datos obtenidos mediante un pHmetro y un colorímetro.

###Marco teórico

El proceso de adición de un volumen medido de una disolución de concentración conocida para que reaccione con una sustancia cuya concentración se desea conocer se llama titulación. La disolución de concentración conocida se llama titulante. Las titulaciones son parte de los métodos volumétricos del análisis químico. Para que un proceso sea susceptible de ser utilizado como una titulación volumétrica debe cumplir con ciertas características:

  -La reacción entre la sustancia de
  concentración desconocida y el titulante debe
  ser sencilla, ya que sirve como base para los
  cálculos.

  -La reacción debe ser rápida, con objeto de
  que la valoración se pueda realizar en poco
  tiempo.

  -La reacción debe ser completa en el
  momento en que se han añadido cantidades
  equivalentes del titulante.

  -Debe existir alguna forma de detectar el
  punto final de la titulación, ya sea por
  métodos colorimétrico o potenciómetro. 

Cabe aclarar que, el método colorimétrico
consiste en añadir un colorante para que se
pueda apreciar a través de un cambio de color
o de vire del indicador ácido-base. Este
cambio deberá apreciarse en el momento en
que se haya añadido una cantidad de titulante
equivalente a la sustancia de concentración
desconocida, es decir, en el punto
estequiométrico de la reacción.
En el método potenciométrico se utiliza el
medidor de pH, que registra directamente los
valores de pH por cada adición de titulante. El
punto final de la titulación se da cuando
después de un cambio brusco de pH medido y
los siguientes valores permanecen constantes
aun cuando se agrega un exceso de titulante.

Al momento de realizar una titulación de tipo ácido-base se lleva a cabo una reacción de neutralización, que según el carácter del ácido se aprecian dos casos:

  1) Titulación ácido débil con base fuerte.

  2) Titulación ácido fuerte con base fuerte.

Debido a lo anterior, es preciso mencionar
que los ácidos y bases fuertes se refieren a
sustancias que tienden a disociarse
completamente con la finalidad de formar
iones en una solución. Mientras que, los
ácidos y bases débiles se ionizan
parcialmente, ya que este proceso es
considerado como reversible. Por tal motivo
es imprescindible conocer la naturaleza del
ácido o base a tratar para conocer su
tendencia de comportamiento al estar en una
solución.

###Discusión de resultados

El desarrollo de esta práctica permitió
mediante la elaboración de dos curvas de
titulación diferenciar la tendencia de la
variación del pH tanto de un ácido débil como
de un aminoácido al añadirle una base fuerte
en ambos casos. De esta manera, se puede
decir al respecto de los datos obtenidos en
cada uno de los procedimientos:

**1) Titulación del ácido débil **

Durante la titulación del ácido acético se
evidenció que a medida en que se iba
añadiendo mL de Hidróxido de Sodio, el pH
iba aumentando de manera directa desde que
estaba muy ácido hasta finalmente llegar a ser
muy base alcanzando su punto de inflexión.

Además, de acuerdo al comportamiento de la
curva de titulación, se observa que la mayoría
de los datos obtenidos del pH tienen una
cercanía de valores, los cuales representan
que el ácido acético en un rango estimado de
pH entre 4 y 6 puede ser una solución
amortiguadora o tampón, debido a que en
dicho intervalo no experimenta a pesar de
seguir aumentando los mL de NaOH cambios
bruscos en esta escala.

Cabe recalcar que mediante el método del
colorímetro se puede observar los cambios
del pH de una forma visual pero no exacta, al
emplear tanto gotas del colorante de la
fenolftaleína, como del uso del papel tornasol
que permite de acuerdo a una escala de color
determinar una cercanía al pH de la solución. 

A continuación, se puede apreciar el cambio
de pH usando dos gotas de fenolftaleína que
fueron agregadas cuando el pH de la solución
era igual a 3,85; que a partir de este instante
cada vez que se agregaba más volumen de
NaOH, la solución se tornaba levemente
como una coloración más morada, hecho que
significa que el ácido acético pasó poco a poco
de ser ácido a ser base.

**2) Titulación de la glicina**

En el caso particular de la glicina, se puede
determinar mediante su curva de titulación
una gran diferencia respecto a la tendencia
que describe la variación del pH de la glicina
al ser comparada con la curva de titulación del
ácido acético, ya que se tratan de dos tipos de
ácidos diferentes, los cuales son un ácido
débil y un ácido fuerte
correspondientemente, y aunque a ambos se
les añadió la misma base fuerte, se puede
decir que el tipo de ácido influye
directamente en el proceso de titulación. 

En la curva de titulación de la glicina se puede
observar que se presentan tres intervalos
diferentes, donde en dos de ellos, el pH tiende
a mantener valores similares, mientras que,
en el intervalo de 0,6mL y 0,7mL de NaOH el
pH sufre un cambio brusco. 

Cabe mencionar que esos dos intervalos,
siendo uno aproximadamente desde 0mL a
0,5mL de NaOH, el pH es muy cercano cada
vez a 4; mientras que el otro intervalo
corresponde desde que el volumen de NaOH
es igual a 0,7mL hasta el volumen final, el pH
se mantiene en un rango entre 9 y 10. Por esta
razón, de no presentar una gran variación en
el pH, dichos intervalos representan
momentos cuando dicha solución que
contiene el aminoácido de la glicina se puede
considerar como una solución tampón,
evidencia de gran importancia biológica al
considerar dicha característica al referirnos a
un aminoácido. 

De la misma manera, al emplear el colorante
de la fenolftaleína se evidencia un cambio de
color en dicha solución, cada vez más intenso
a medida que la solución pasa de ser ácida o
ser más base. 

###Conclusiones

- Mediante este laboratorio se pudo
evidenciar la variación que sufre el pH al
realizar un proceso de titulación con dos
tipos de ácidos junto con una base fuerte.

- El pH se puede determinar mediante el uso
de un pHmetro de manera exacta o
usando el método visual del colorímetro
con un colorante o tiras de papel tornasol. 

- Se analizó que, aunque las curvas de
titulación tengan una tendencia diferente,
existen intervalos donde dichas soluciones
se pueden considerar como soluciones
tampón, debido a que el pH trata de
mantener valores similares, aunque el
volumen siga aumentando. 

- Las soluciones tampón tienen una gran
importancia biológica, ya que al tender a
mantener relativamente estable el pH
ante sus condiciones juegan un papel
importante para evitar posibles daños en
los organismos vivos.
