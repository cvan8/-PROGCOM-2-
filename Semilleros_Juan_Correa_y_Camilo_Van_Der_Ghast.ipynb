# **#1. Solución  al problema de ruteo de vehículos de flota heterogénea con transbordo  lateral  mediante la metaheurística de algoritmo genético**

WhatsApp Image 2022-10-06 at 10.50.20 AM.jpeg


Grupo de investigación : OPALO

Integrantes: Ingrith Marcela Mendoza y Carlos Enersto Mendoza 

Línea de acción: 

Resumen:

En este semillero se busca crear una solución al problema de ruteo de vehículos de flota heterogénea con trasbordo lateral,  en otras palabras se enfrenta a un problema de matemático y de optimización que plantea la asignación adecuada de rutas para la distribución de mercancía desde un punto de origen a diferentes clientes utilizando una flota de vehículos. Esta tarea se ve en dificultad en gran variedad de casos, gracias a variables como leyes y restricciones de tránsito, accidente y tráfico, falta de rutas y vías etc….

La solución que plantea este semillero hace referencia a la creación de un modelo que pueda ser empleado por las diferentes en empresa para dar solución a este inconveniente, para desarrollar dicho modelo el grupo emplea un algoritmo genético, un algoritmo es una serie de pasos organizados que describe el proceso que se debe seguir para dar solución a un problema específico, y un algoritmo genético tiene la características de simular situaciones e ir evolucionando a dichos sucesos para formular esos pasos y hallar la solución, de manera más resumida es una inteligencia artificial donde su funcionamiento principal gira alrededor de la adaptación a dichas situaciones.

El semillero apunta al resultado de crear y validar dicho algoritmo como la respuesta para la problemática  ruteo de vehículos de flota heterogénea con trasbordo lateral, donde esté en diferentes situaciones y teniendo en cuenta diferentes restricciones y reglas de tránsito pueda dar una solución factible en cuanta a la ruta que debe ser tomada para llegar al punto del cliente.



#2. Evaluación del ruido ambiental en el campus  de la universidad de Santander con un módulo electrónico de bajo costo en tiempo real.

WhatsApp Image 2022-10-06 at 10.50.13 AM.jpeg


Grupo de investigación: GAIA (Grupo Ambiental de Investigación Aplicada).

Integrantes: Walter Pardavé Livia.

Línea de acción: Investigación Calidad del aire, tratamiento de residuos industriales y tecnologías sostenibles.

Resumen:

El propósito del proyecto presentado por este semillero es el de evaluar el nivel de ruido ambiental en el campus de la Universidad de Santander a tiempo real empleando un módulo electrónico de bajo costo. Aunque no se escucha al nivel de la contaminación de aguas o tierra la contaminación sonora también es un riesgo para el medio ambiente, se entiende por contaminación sonora o acústica al exceso de sonido que altera las condiciones normales del ambiente en una determinada zona. Y si bien el ruido no se acumula, traslada o perdura en el tiempo como sí ocurre con las otras formas de contaminación mencionadas anteriormente , la contaminación sonora puede causar grandes daños en la calidad de vida de las personas si no se controla bien o adecuadamente.

Para la aplicación de este proyecto se escogieron diferentes lugares de monitoreo a través del Campus de la Universidad de Santander, dichos puntos fueron escogidos teniendo en cuenta las fuentes de ruido más comunes en el área. La información fue presentada en forma de una tabla de barra donde se comparaba los db del sonido osea los niveles de intensidad sonora vs el tiempo del día, los resultados obtenidos por el grupo muestran que los niveles durante el horario diurno, dia habil y festivo en los tres puntos de monitoreo establecidos presentan niveles de presión sonora aceptables al límite máximo permisible según la normativa, el cual considera entre los 70 dB para zonas de constante ruido como lo era la universidad y prohíbe las exposicion a presion de 115 db o mayor.

#3. Sistema de sincronía entre equipos de monitoreo de variables fisiológicas.

WhatsApp Image 2022-10-06 at 10.50.18 AM.jpeg

Grupo de investigación: SEBIUNAB.

Integrantes: J.Angulo, P. Jiménez, V. Valzania.

Linea de acción: Neurociencias, economía y equipos biomédicos.

Resumen:

En este semillero se busca desarrollar un sistema de sincronización en equipos de monitoreo para la caracterización de variables fisiológicas.
Para la realización de este sistema, ellos utilizarán el software de MATLAB, un sistema de cómputo numérico que ofrece un entorno de desarrollo integrado con un lenguaje de programación propio.

Para la realización de este proyecto, se identificará principalmente el funcionamiento, procesamiento de señales y el manejo de los equipos biomédicos. En este caso, se utilizará como base la electrocardiografía y la electroencefalografía. Se utilizarán estos dos equipos para poder analizar las variables fisiológicas presentadas en los humanos y se realizará la debida experimentación de la unión de estos dos equipos a través de MATLAB.
El semillero apunta al resultado de poder unir los diversos equipos con el fin de facilitar la identificación de diversas características que presentan estos acerca de la lectura del cuerpo humano, lo cual daría una lectura más precisa y rápida para que el correspondiente encargado de leer lo que arroja la máquina, pueda dar un servicio más eficaz.

#6. Recursos didácticos desde la arquitectura que identifican la flora nativa en el área metropolitana de Bucaramanga.

WhatsApp Image 2022-10-06 at 11.06.31 PM.jpeg


Grupo de investigación: GINVEARQUI - FACARQ

Integrantes: Universidad Santo Tomás,

Línea de acción: Planificación y gestión del territorio.

Resumen:

El tema principal que gira alrededor de la investigación y desarrollo del semillero es la biodiversidad de la flora , en el caso de esta iniciativa presentada  se busca generar un paquete de recursos didácticos para la representación gráfica de la vegetación en planos de arquitectura.

El problema gira alrededor de la que la falta de material gráfico para la representación de la flora en construcción y remodelación, carece de la importancia que debería tener, ejemplo de esto es que es uno de los últimos temas y que menos recibe acompañamiento en el dibujo de arquitectura para los estudiantes.

Por eso con esta iniciativa se busca tener tres objetivos específicos primero la preparar la digitalización de especies de floras seleccionadas en planos de arquitectura, el siguiente hace referencia a el diseño de una plantilla de dibujo a nivel industrial para ser integrados com un producto y por último es difundir los recurso generados de forma virtual para dar importancia sobre la flora nativa en el área metropolitana  de bucaramanga.

Los resultados que se esperan obtener sobre esta iniciativa son los siguiente, se espera gráficos digitalizables de al menos 3 especies nativas de flora de bucaramanga, disposición de plantillas para la representación del dibujo a mano alzada y por últimos un video explicativo que dé instrucciones sobre el uso de los recursos construidos y generados en esta propuesta.
